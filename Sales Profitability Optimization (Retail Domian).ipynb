```python
#  to find out the median reviwe rating withn each category and will replace that with the missing value
# if clothing item missing its reviwe rating it gets replace with the median reviwe rating of clothing item not a footware item
df ['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# for evry uppercase in category names convert lower case with under score becouse of coding
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
# to renmae  Purchase Amount (USD) 
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
df.columns
```

```python
# create a column age_group
labels = ['Young Adult','Adult','Middle-agedd','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
df[['age','age_group']].head(10)
```
